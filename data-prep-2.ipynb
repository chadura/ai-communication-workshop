```python
fs = "demos/sentiment-api/data/sentiment_sample.csv"

with open(fs, 'r') as inp:
    lines = inp.readlines()
```

```python
texts = []
labels = []

for line in lines[0:45]:
    cols = line.replace('.,', '\t')
    cols = cols.replace('!,', '\t')
    cols = cols.replace(',\s', '\t')
    out = cols.split('\t')
    texts.append(out[0])
    labels.append(out[-1].strip())
```

```python
df = pd.DataFrame([texts, labels])
```
